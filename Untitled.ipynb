## 1. Load Libraries

## 5. Prepare Data

In this section, I will do following:
- Define **X** and **y** (where X - input variables and y - target or dependent variable)
- Split dataset into train and test set
- Normalize features

```python
X = df.iloc[:,0:13]
y = df['MEDV']
```

```python
# normalize features
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=1)
```

```python
print("Train data shape: ",X_train.shape, y_train.shape)
print("Test data shape: ",X_test.shape, y_test.shape)
```

## 6. Model Training and Evaluation

In this section, I will train following algorithms:
- Linear Regression
- Ridge Regression
- Lasso Regression
- KNeighborsRegressor
- DecisionTreeRegressor
- SVR

and evaluate these algorithms using KFold cross-validation and mean_squared_error metric.

```python
# create instances of each model
models = []
models.append(('LR', LinearRegression()))
models.append(('RIDGE', Ridge()))
models.append(('LASSO', Lasso()))
models.append(('KNN', KNeighborsRegressor()))
models.append(('CART', DecisionTreeRegressor()))
models.append(('SVR', SVR(gamma='auto')))
```

```python
# train and evaluate each algorithm
results = []
names = []
for name, model in models:
    kfold = KFold(n_splits = 5, random_state=1)
    cv_results = cross_val_score(model, X_train, y_train, cv = kfold, scoring='neg_mean_squared_error')
    results.append(cv_results)
    names.append(name)
    print("%s: %f" % (name, cv_results.mean()))
```

## 7. Improve Results with Tunning

```python
# choose best value of k for KNN
k_values = np.array([1,3,5,7,9,11,13,15])
param_grid = dict(n_neighbors=k_values)
model = KNeighborsRegressor()
kfold = KFold(n_splits=5, random_state=1)
grid = GridSearchCV(estimator=model, param_grid=param_grid, scoring='neg_mean_squared_error', cv=kfold)
grid_result = grid.fit(X_train,y_train)
```

```python
# display mean scores as well as best performing value for KNN
print("Best Score: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
params = grid_result.cv_results_['params']

print()
for mean, param in zip(means, params):
    print("Score: %f using %s" % (mean, param))
```

## 8. Ensemble Methods

Another way that we can improve the performance of algorithms on this problem is by using ensemble methods.

In this section, I will evaluate the following ensembe machine learning algorithms:
- RandomForests
- AdaBoost
- GradientBoosting

```python
# ensembles
ensembles = []
ensembles.append(('AB', AdaBoostRegressor()))
ensembles.append(('GBM', GradientBoostingRegressor()))
ensembles.append(('RF', RandomForestRegressor(n_estimators=100)))

results = []
names = []
for name, model in ensembles:
    kfold =KFold(n_splits=5, random_state=1)
    cv_results = cross_val_score(model, X_train, y_train, scoring='neg_mean_squared_error', cv=kfold)
    results.append(cv_results)
    names.append(name)
    print("%s: %f (%f)" % (name, cv_results.mean(), cv_results.std()))
```

As you can see that RandomForestRegressor gives us the best results. Now I wil further tune this model and finalize it.

## 9. Tunning Ensemble Methods

```python
param_grid = dict(n_estimators=np.array([50,100,150,200,250,300,350,400]))
model = RandomForestRegressor(random_state=1)
kfold = KFold(n_splits=5, random_state=1)
grid = GridSearchCV(estimator=model, param_grid=param_grid, scoring='neg_mean_squared_error', cv=kfold)
grid_result = grid.fit(X_train, y_train)
```

```python
# show results from GridSearchCV
print("Best Score: %f using %r" % (grid_result.best_score_, grid_result.best_params_))

means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']

print()
for mean, std, param in zip(means, stds, params):
    print("Score: %f, Std: %f using %s" %(mean, std, param))
```

## 10. Finalize Model

As we know that RandomForest gives us best results. we have also tunned this model.
Now let's retrain this model on training set with best parameters and make prediction on test set.

```python
model = RandomForestRegressor(n_estimators=100)
model.fit(X_train,y_train)
```

```python
# make prediction
preds = model.predict(X_test)
mean_squared_error(preds,y_test)
```

We can see that test error is very close to training error, So our model is giving good results.
